```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

## Carregamento do Dataset

O conjunto de dados foi carregado a partir de um link externo.

```python
url = "https://pastebin.com/raw/J6hYWEUy"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
data = StringIO(response.text)
df = pd.read_csv(data)

print("Primeiras linhas do dataset:")
display(df.head())
```

## 1. Análise Exploratória

Nesta etapa foi feita uma análise inicial da estrutura dos dados, dos tipos de variáveis e das estatísticas principais.

```python
print("Dimensão do dataset (linhas, colunas):")
print(df.shape)

print("\nNomes das colunas:")
print(df.columns)

print("\nInformações gerais:")
df.info()

print("\nEstatísticas descritivas das variáveis numéricas:")
display(df.describe())

print("\nEstatísticas gerais incluindo variáveis categóricas:")
display(df.describe(include="all"))
```

O dataset possui informações de funcionários, incluindo dados pessoais, profissionais e indicadores como satisfação, desempenho e turnover.

### Distribuição das variáveis categóricas

As variáveis categóricas foram analisadas para verificar a frequência de cada categoria e identificar possíveis problemas de padronização.

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    print(f"\nColuna: {col}")
    print(df[col].value_counts(dropna=False))
```

## 2. Identificação de Problemas de Qualidade

Nesta parte foram verificados valores ausentes, inconsistências e outliers.

```python
print("\nValores ausentes por coluna:")
display(df.isnull().sum())

print("\nPercentual de valores ausentes por coluna:")
display((df.isnull().sum() / len(df) * 100).sort_values(ascending=False))
```

Foi feita a verificação de valores ausentes para identificar colunas que precisavam de tratamento antes das análises finais.

```python
print("\nValores únicos em colunas com possível inconsistência:")
for col in ["Genero", "Departamento", "Cargo", "Formacao"]:
    print(f"\n{col}:")
    print(df[col].unique())
```
